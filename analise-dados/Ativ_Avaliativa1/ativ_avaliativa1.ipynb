```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
 
```

Questão 1 \
Usando o dataset 'nba2k.csv', gere um gráfico que exiba o valor do salário dos 10 jogadores mais altos da NBA.

```python
df= pd.read_csv("nba2k.csv", encoding ='latin-1')
df2 = df.nlargest(10, 'height')
df3 = df2.sort_values(by='salary', ascending=False)

plt.rcParams['figure.figsize'] = [12,6]

df3.plot(x='full_name', y='salary', kind='bar', stacked=False)
plt.title('Salario dos 10 jogadores mais altos da NBA')
plt.xlabel('Jogadores')
plt.ylabel('Salario')
plt.show()
```

Questão 2 \
Gere um gráfico que exiba o salário dos 10 jogadores estrangeiros mais bem pagos da NBA.

```python
df_gringo = df[df['country'] != 'USA']
df_gringo2 = df_gringo.nlargest(10, 'salary')
df_gringo3 = df_gringo2.sort_values(by='salary', ascending=False)

plt.rcParams['figure.figsize'] = [12,6]

df_gringo3.plot(x='full_name', y='salary', kind='bar', stacked=False)
plt.title('10 estrangeiros mais bem pagos da NBA')
plt.xlabel('Jogadores')
plt.ylabel('Salario')
plt.show()
```

Questão 3 \
Gere dois gráficos de pizza lado a lado que exibam: \
A) o percentual de jogadores americanos vs jogadores estrangeiros na NBA; \
B) o valor total pago a jogadores americanos vs jogadores estrangeiros na NBA.

```python
df_usa = df[df['country'] == 'USA']
usa_players = len(df_usa)
gringo_players = len(df_gringo)

# get the salary of all persons that not american
gringo_salary_all = sum(df_gringo['salary'])
usa_salary_all = sum(df_usa['salary'])

labels = 'Estrangeiros', 'Americanos'
plt.rcParams['figure.figsize'] = [12,8]

plt.subplot(1,2,1)
plt.pie([usa_players, gringo_players], labels= labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Quantidade de americanos vs estrangeiros da NBA')

plt.subplot(1,2,2)
plt.pie([usa_salary_all, gringo_salary_all], labels= labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Valores pagos: americanos vs estrangeiros na NBA')

plt.tight_layout()
plt.show()

```

Questão 4 \
Utilizando o dataset "world_population.csv", gere um gráfico que exiba a evolução da população por continente nas últimas 6 décadas (1970 a 2020).

```python
df_world = pd.read_csv("world_population.csv", encoding ='latin-1')

df_world['Continent'].unique()

df_world_filtered = df_world.filter(items=['1970 Population', '1980 Population', '1990 Population', '2000 Population', '2010 Population', '20 Population'])

plt.rcParams["figure.figsize"] = [10, 8]

sns.set_context("notebook", font_scale=1.5, rc={"fontsize": 14, "axes.titlesize": 16, "axes.labelsize": 12})

df_world_filtered.plot(kind='bar', stacked=False)

plt.title("Evolução da população por continente nas ultimas 6 decadas")
plt.ylabel("População em bilhoes")
plt.xlabel("Continente")
plt.show()
```

Questão 5 \
Utilizando o dataset "hapiness2021.csv, gere dois gráficos de dispersão lado a lado comparando a relação do índice de felicidade (Ladder score) e a renda per capita (Logged GDP per capita) nos países da África (Sub-Saharan Africa) e da Europa Ocidental (Western Europe).

```python
df_hap = pd.read_csv("hapiness2021.csv", encoding ='latin-1')
#Logged GDP per capita - y
# Ladder score	 - x

df_europe = df_hap[df_hap['Regional indicator'] == 'Western Europe']
df_africa = df_hap[df_hap['Regional indicator'] == 'Sub-Saharan Africa']

plt.rcParams["figure.figsize"] = [12, 8]

plt.subplot(1,2,1)

sns.regplot( x="Ladder score", y="Logged GDP per capita", data=df_europe)

plt.title("Indice de felicidade e renda per capita (Europa)")
plt.xlabel("Indice de Felicidade")
plt.ylabel("Renda per capita")


plt.subplot(1,2,2)

sns.regplot( x="Ladder score", y="Logged GDP per capita", data=df_africa)

plt.title("Indice de felicidade e renda per capita (Africa)")
plt.xlabel("Indice de Felicidade")
plt.ylabel("Renda per capita")

plt.tight_layout()
plt.show()


```
