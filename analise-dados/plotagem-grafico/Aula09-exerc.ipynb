Questão 1 \
Utilizando o dataset 'tips' crie um gráfico de barras mostrando o faturamento total do restaurante por dia, exibindo o total gasto por homens e mulheres.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
fat_dia = pd.crosstab(df['day'], df['sex'], values=df['total_bill'], aggfunc=np.sum)
fat_dia
```

```python
plt.rcParams["figure.figsize"] = [6,4]
sns.set_context("notebook", font_scale=1.5, rc={"font.size":14,"axes.titlesize":16,"axes.labelsize":12})
fat_dia.plot(kind="bar",stacked=False)
plt.title("Faturamento Homem x Mulher")
plt.ylabel("Faturamento")
plt.xlabel("Dia da semana")
plt.show()
```

Questão 2 \
Elabre um gráfico de pizza que mostre a divisão do faturamento entre homens e mulheres.

```python
df_homens = df[df['sex']=='Male']
df_mulheres = df[df['sex']=='Female']

sizes = [df_homens['total_bill'].sum(), df_mulheres['total_bill'].sum()]
labels = ['Homens', 'Mulheres']
sns.set_style("darkgrid")
plt.pie(sizes, labels = labels)
plt.title('Divisão do faturamento entre homens e mulheres')
plt.axis('equal')
plt.show()
```

Questão 3 \
Elabore um gráfico de dispersão que mostre a relação entre o número de pessoas por mesa e o valor total da conta.

```python
df2 = df[df['total_bill'] < 45] 
sns.regplot(x="size", y="total_bill", fit_reg=True, data=df2)
plt.title("Relação entre o número de pessoas e o valor da conta")
plt.xlabel("Pessoas na mesa")
plt.ylabel("Valor da conta")
plt.show()
```
